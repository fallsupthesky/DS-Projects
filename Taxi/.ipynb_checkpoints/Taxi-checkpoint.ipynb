# 3. Обучение

Создадим отдельную функцию для добавления признаков в датафрейм 'make_features' и функцию разделения на тренировочную, валидационную, тестовую выборки 'split':

```python
def make_features(data, max_lag, rolling_mean_size):
    data_features = data.copy()
    data_features['month'] = data.index.month
    data_features['day'] = data.index.day
    data_features['dayofweek'] = data.index.dayofweek
    data_features['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data_features['lag_{}'.format(lag)] = data_features['num_orders'].shift(lag)
        
    data_features['rolling_mean'] = data_features['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data_features
```

```python
def split(data):
    train, temp = train_test_split(data_features, shuffle=False, test_size=0.2, random_state=1234)
    valid, test = train_test_split(temp, shuffle=False, test_size=0.5, random_state=1234)
    train = train.dropna()
    
    features_train = train.drop('num_orders', axis=1)
    features_valid = valid.drop('num_orders', axis=1)
    features_test = test.drop('num_orders', axis=1)

    target_train = train['num_orders']
    target_valid = valid['num_orders']
    target_test = test['num_orders']
    
    return features_train, features_valid, features_test, target_train, target_valid, target_test
```

Создадим таблицу с результатами предсказаний:

```python
model_results = pd.DataFrame(
    index=
        [
        'Linear_regression',
        'Decision_tree',
        'Random_forest',
        'LightGBM',
        'CatBoost'
        ],
    
    columns=
        [
        'rmse_valid',
        'rmse_test',
        ]
                            )
```

Для поиска оптимального кол.-ва лагов и скользящего среднего воспользуемся графиками автокорреляции и частичной автокорреляции.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
ts = data['num_orders']
plot_pacf(ts, lags=50).set_size_inches(12, 6)
plt.show()
```

Запишем несколько значимых значений частичной автокорреляции, после которых большинство лагов являются незначимыми: 2, 8, 12, 14, 16, 18, 28. 

Построим график автокорреляции:

```python
plot_acf(ts, lags=50).set_size_inches(12, 4)
plt.show()
```

Чтобы избавиться об большого числа значимых лагов в автокорреляции, найдем разность первого порядка и построим новый график автокорреляции:

```python
ts_diff = ts - ts.shift(1)
ts_diff = ts_diff.dropna()
```

```python
plot_acf(ts_diff, lags=50).set_size_inches(12, 6)
plt.show()
```

Возьмем значимые лаги 1, 21, 24, 28, больше которых лежит большинство незначимых.

Найдем наилучшее значение кол.-ва лагов и скользящего среднего перебором найденных значений в цикле с линейной регрессией:

```python
%%time
results = []

for lag in [2, 8, 12, 14, 16, 18, 28]:
    for rolling in [1, 21, 24, 28]:
        data_features = make_features(data, lag, rolling)
        features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)

        model_linear = LinearRegression()
        model_linear.fit(features_train, target_train)
        predictions_valid_linear = model_linear.predict(features_valid)

        rmse_valid_linear = mean_squared_error(target_valid, predictions_valid_linear)**0.5
        result = lag, rolling, rmse_valid_linear
        results.append(result)

df_linear = pd.DataFrame(results, columns=['lag', 'rolling', 'rmse_valid'])
print('Наименьшее значение RMSE получено при следующих параметрах:')
df_linear.loc[df_linear['rmse_valid'].idxmin()]
```

Запишем полученные оптимальные значения в таблицу:

```python
best_lag_linear = df_linear.loc[df_linear['rmse_valid'].idxmin(), 'lag']
best_rolling_linear = df_linear.loc[df_linear['rmse_valid'].idxmin(), 'rolling']
model_results.loc['Linear_regression', 'rmse_valid'] = round(df_linear.loc[df_linear['rmse_valid'].idxmin(), 'rmse_valid'], 2)
```

В дальнейшем будем использовать кол.-во лагов 28, скользящее среднее 24.

Создадим модели и найдем значения RMSE на валидационной выборке.

```python
# код ревьювера
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
ts = data['num_orders']
plot_pacf(ts, alpha=None).set_size_inches(12, 4)
plt.show()

plot_acf(ts, alpha=None).set_size_inches(12, 4)
plt.show()
```

- Decision Tree

Подберем оптимальные гиперпараметры:

```python
%%time

tree_results = []
lag = 28
rolling = 24

for depth in range(2,11,2):
    for samp_split in range(2,5):
        for leaf in range(1,6):
        
            data_features = make_features(data, lag, rolling)
            features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)

            model_tree = DecisionTreeRegressor(
                max_depth=depth, 
                min_samples_split=samp_split, 
                min_samples_leaf=leaf,
                random_state=1234
                ) 
            model_tree.fit(features_train, target_train)
            predictions_valid_tree = model_tree.predict(features_valid)

            rmse_valid_tree = mean_squared_error(target_valid, predictions_valid_tree)**0.5
            result = lag, rolling, depth, samp_split, leaf, rmse_valid_tree
            tree_results.append(result)
            
df_tree = pd.DataFrame(tree_results, columns=['lag', 'rolling', 'depth', 'samp_split', 'leaf', 'rmse_valid'])
print('Наименьшее значение RMSE получено при следующих параметрах:')
df_tree.loc[df_tree['rmse_valid'].idxmin()]
```

```python
best_lag_tree = 28
best_rolling_tree = 24
best_depth_tree = df_tree.loc[df_tree['rmse_valid'].idxmin(), 'depth']
best_split_tree = df_tree.loc[df_tree['rmse_valid'].idxmin(), 'samp_split']
best_leaf_tree = df_tree.loc[df_tree['rmse_valid'].idxmin(), 'leaf']

model_results.loc['Decision_tree', 'rmse_valid'] = round(df_tree.loc[df_tree['rmse_valid'].idxmin(), 'rmse_valid'], 2)
```

- Random Forest

```python
%%time

forest_results = []

for depth in [14]:
    for estimator in [100]:
        for samp_split in range(2,4):
            for leaf in range(1,3):

                data_features = make_features(data, lag, rolling)
                features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)

                model_forest = RandomForestRegressor(
                    max_depth=depth,
                    n_estimators=estimator,
                    random_state=1234,
                    n_jobs=-1
                    ) 
                model_forest.fit(features_train, target_train)
                predictions_valid_forest = model_forest.predict(features_valid)

                rmse_valid_forest = mean_squared_error(target_valid, predictions_valid_forest)**0.5
                result = lag, rolling, depth, estimator, samp_split, leaf, rmse_valid_forest # depth, estimator,
                forest_results.append(result)
            
df_forest = pd.DataFrame(forest_results, columns=['lag', 'rolling', 'depth', 'estimator', 'samp_split', 'leaf', 'rmse_valid'])
print('Наименьшее значение RMSE получено при следующих параметрах:')
df_forest.loc[df_forest['rmse_valid'].idxmin()]  
```

```python
best_lag_forest = 28
best_rolling_forest = 24
best_depth_forest = df_forest.loc[df_forest['rmse_valid'].idxmin(), 'depth']
best_estimator_forest = df_forest.loc[df_forest['rmse_valid'].idxmin(), 'samp_split']
best_split_forest = df_forest.loc[df_forest['rmse_valid'].idxmin(), 'samp_split']
best_leaf_forest = df_forest.loc[df_forest['rmse_valid'].idxmin(), 'leaf']

model_results.loc['Random_forest', 'rmse_valid'] = round(df_forest.loc[df_forest['rmse_valid'].idxmin(), 'rmse_valid'], 2)
```

- LightGBM

```python
%%time

lgb_results = []

for depth in range(10,20):
    for estimator in [100]:
        data_features = make_features(data, lag, rolling)
        features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)
        categorical = ['month', 'day', 'dayofweek', 'hour']                                                                                   
        lgb_train = lgb.Dataset(features_train, target_train, categorical_feature = categorical)
        lgb_valid = lgb.Dataset(features_valid, target_valid, categorical_feature = categorical)
        lgb_test = lgb.Dataset(features_test, target_test, categorical_feature = categorical)

        params = {
            'objective': 'regression',
            'boosting': 'gbdt',
            'metric': 'rmse',
            'n_estimators': estimator,
            'max_depth': depth,
            'verbose': -1,
            'random_state': 1234,
            'n_jobs': -1
            }

        gbm = lgb.train(
            params, 
            lgb_train, 
            valid_sets=lgb_valid, 
            verbose_eval=False
            )
        
        predictions_valid_lgb = gbm.predict(features_valid, num_iterationl=gbm.best_iteration)
        rmse_valid_lgb = mean_squared_error(target_valid, predictions_valid_lgb)**0.5
        result = lag, rolling, depth, estimator, rmse_valid_lgb
        lgb_results.append(result)
```

```python
df_lgb = pd.DataFrame(lgb_results, columns=['lag', 'rolling', 'depth', 'estimator', 'rmse_valid'])
print('Наименьшее значение RMSE получено при следующих параметрах:')
df_lgb.loc[df_lgb['rmse_valid'].idxmin()]
```

```python
best_lag_lgb = 28
best_rolling_lgb = 24
best_depth_lgb = df_lgb.loc[df_lgb['rmse_valid'].idxmin(), 'depth']
best_estimator_lgb = df_lgb.loc[df_lgb['rmse_valid'].idxmin(), 'estimator']

model_results.loc['LightGBM', 'rmse_valid'] = round(df_lgb.loc[df_lgb['rmse_valid'].idxmin(), 'rmse_valid'], 2)
```

- CatBoost

```python
%%time

catboost_results = []

for depth in [5,7]:
        
    data_features = make_features(data, lag, rolling)
    features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)

    model_catboost = CatBoostRegressor(
                    depth=depth, 
                    loss_function='RMSE',
                    verbose=False,
                    random_state=1234,
                    )          
    model_catboost.fit(features_train, target_train)
    predictions_valid_catboost = model_catboost.predict(features_valid)
        
    rmse_valid_catboost = mean_squared_error(target_valid, predictions_valid_catboost)**0.5
    result = lag, rolling, depth, rmse_valid_catboost
    catboost_results.append(result)
        
df_catboost = pd.DataFrame(catboost_results, columns=['lag', 'rolling', 'depth', 'rmse_valid'])
print('Наименьшее значение RMSE получено при следующих параметрах:')
df_catboost.loc[df_catboost['rmse_valid'].idxmin()]
```

```python
best_lag_catboost = 28
best_rolling_catboost = 24
best_depth_catboost = df_catboost.loc[df_catboost['rmse_valid'].idxmin(), 'depth']

model_results.loc['CatBoost', 'rmse_valid'] = round(df_catboost.loc[df_catboost['rmse_valid'].idxmin(), 'rmse_valid'], 2)
```

- LightGBM

```python
data_features = make_features(data, best_lag_lgb, best_rolling_lgb)
features_train, features_valid, features_test, target_train, target_valid, target_test = split(data_features)
categorical = ['month', 'day', 'dayofweek', 'hour']                                                                                    
lgb_train = lgb.Dataset(features_train, target_train, categorical_feature = categorical)
lgb_valid = lgb.Dataset(features_valid, target_valid, categorical_feature = categorical)
lgb_test = lgb.Dataset(features_test, target_test, )
            
params = {
    'objective': 'regression',
    'boosting': 'gbdt',
    'metric': 'rmse',
    'max_depth': best_depth_lgb,
    'n_estimators': best_estimator_lgb,
    'verbose': -1,
    'random_state':1234,
    'n_jobs': -1
            }           
     
gbm = lgb.train(
    params, 
    lgb_train, 
    valid_sets=lgb_valid, 
    verbose_eval=False
                )

predictions_test_lgb = gbm.predict(features_test, num_iteration=gbm.best_iteration)
rmse_test_lgb = mean_squared_error(target_test, predictions_test_lgb)**0.5
model_results.loc['LightGBM', 'rmse_test'] = round(rmse_test_lgb, 2)
print('RMSE LightGBM на тестовых данных:', round(rmse_test_lgb, 2))
```
